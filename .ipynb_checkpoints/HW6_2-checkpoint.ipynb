## Overfitting and Deterministic Noise
### Question 1
Deterministic noise depends on $\mathcal H$, as some models approximate $f$ better than others. Assume that $\mathcal H' \subset \mathcal H$ and that $f$ is fixed. In general (but not necessarily in all cases), if we use $\mathcal H'$ instead of $\mathcal H$, how does deterministic noise
behave?

**Answer**:  
Deterministic noise is when the given hypothesis can not model the target function properly, i.e. there is a "*gap*" from the hypothesis to the actual target function. Since $f$ is fixed we can concentrate only on the hypothesis set. Given that $\mathcal H'$ is a subset of $\mathcal H$ it is a smaller model and its in-sample error will be higher than for $\mathcal H$ since it has less degrees of freedom to fit the data. In other words the "*gap*" is larger with $\mathcal H'$ than with $\mathcal H$. Hence, the derministic noise should be higher.   

The answer is (b).

## Regularization with Weight Decay
$\newcommand{\R}{\mathbb{R}}$
For the following problems we use the data provided in the files [train data](http://work.caltech.edu/data/in.dta) and 
[test data](http://work.caltech.edu/data/out.dta) as a training and test set respectively. Each line of the files corresponds to a two-dimensional input $x = (x_1, x_2)$, so that $\mathcal X = \R^2$, followed by the corresponding
label from $\mathcal Y = \{-1, 1\}$. We are going to apply Linear Regression with a non-linear
transformation for classification. The nonlinear transformation is given by

$$
\Phi(x_1, x_2) = (1, x_1, x_2, x^2_1, x^2_2, x_1x_2, |x_1-x_2|, |x_1+x_2|)
$$

Recall that the classification error is defined as the fraction of misclassified points.
### Question 2
Run Linear Regression on the training set after performing the non-linear transformation. What values are closest (in Euclidean distance) to the in-sample and out-of-sample classification errors, respectively?  

**Answer**:  

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

First we import the given data, both train and test samples.

```python
# Import data
train = np.loadtxt('Data/in.dta')
test = np.loadtxt('Data/out.dta')
N = train.shape[0] # number of examples in train set
Nt = test.shape[0] # number of examples in test set
```

Let's check the structure of data to see if everything is imported OK.

```python
print('The dimensions of the train data sample is :',train.shape)
print('The dimensions of the test data sample is :',test.shape)
```

Let's rename the data sets and transform the input variables.

```python
Z = np.column_stack([np.ones(train.shape[0]), train[:,:2], #x1,x2
                     train[:,:2]**2, #x1^2, x2^2
                     train[:,0]*train[:,1], #x1x2
                    np.absolute(train[:,0] - train[:,1]), #|x1-x2|
                    np.absolute(train[:,0] + train[:,1])]) #|x1+x2|
```

```python
Zt = np.column_stack([np.ones(test.shape[0]), test[:,:2], 
                     test[:,:2]**2, test[:,0]*test[:,1],
                    np.absolute(test[:,0] - test[:,1]),
                    np.absolute(test[:,0] + test[:,1])])
```

```python
X = Z
y = train[:,2]
Xt = Zt
yt = test[:,2]
m = Z.shape[1] # number of parameters in train
```

```python
# 2. Linear regression
def linregwd(k=None):
    if k:
        l = np.power(10,float(k))
    else:
        l = 0
    # Compute the weights
    reg_term = l*np.eye(m)
    w = np.dot(np.dot(np.linalg.inv(np.dot(X.T,X) + reg_term),X.T),y)
    
    # Estimate hypothesis
    g = np.where(np.dot(X,w) >= 0, 1, -1)
        
    # Compute in-sample error
    Error = g - y
    Ein = np.count_nonzero(Error)/N
    
    gt = np.where(np.dot(Xt,w) >= 0, 1, -1)
    Error_t = gt - yt
    Eout = np.count_nonzero(Error_t)/Nt
       
    lr_out = {'w': w,
              'Ein': Ein,
              'Eout': Eout}
    
    return lr_out
```

```python
lr_out = linregwd()
wl0 = lr_out['w']
lr_out
```

The values are closest to 0.03 and 0.08 so the answer is (a).

## Regularization for Polynomials
Polynomial models can be viewed as linear models in a space $\mathcal Z$, under a nonlinear transform $\Phi :\mathcal X \to \mathcal Z$. Here, $\Phi$ transforms the scalar $x$ into a vector $\mathbf z$ of Legendre polynomials, $\mathbf z = (1, L_1(x), L_2(x), \dots , L_Q(x))$. Our hypothesis set will be expressed as a linear combination of these polynomials,
$$
\mathcal H_Q = \bigg\{ h | h(x)=\mathbf w^T \mathcal z = \sum_{q=0}^{Q} w_qL_q(x) \bigg\}
$$

where $L_0(x)=1$

## Neural Networks
### Question 8
A fully connected Neural Network has $L = 2; d(0) = 5; d(1) = 3; d(2) = 1$. If only products of the form $w^{(l)}_{ij}x^{(l-1)}_i$, $w^{(l)}_{ij}\delta^{(l)}_j$, and $x^{(l-1)}_i \delta^{(l)}_j$ count as operations (even for $x^{(l-1)}_0=1$), without counting anything else, which of the following is the **closest** to the total number of operations in a single iteration of backpropagation (using SGD on one data point)?

**Answer**:
The network has two layers and we can visualize it as follows:
<img src="P8_NN.png" alt="Drawing" style="width: 450px;"/>

Recall the following equations for forward and backward propagation:  

$$
s^{(l)}_j = \sum^{d^{(l-1)}}_{i=0} w^{(l)}_{ij}x^{(l-1)}_i \\
x^{(l)}_j = \phi(s^{(l)}_j)
$$

$$
\frac{\partial e(\mathbf w)}{\partial w_{ij}}=\frac{\partial e(\mathbf w)}{\partial s^{(l)}_j} \times \frac{\partial s^{(l)}_j}{\partial w^{(l)}_{ij}} = \delta^{(l)}_jx^{(l-1)}_i
$$

$$
\delta^{(l-1)}_j = (1-(x^{(-1)}_i)^2)\sum^{j=1}_{d^{(l)}}w^{(l)}_{ij}\delta^{(l)}_j
$$
For the complete network we only need to count the operations stated in the problem.

For forward propagation we have that the dimension of $x^{(0)}$ is $(6,1)$ since there are 5 input values and one for the $x^{(0)}_0$. Recall that the number of nodes is computed using the following $d^{(l)}+1$ except for the $L$ layer where the nodes equal to $d^{(l)}$. Then for the weights we have $\mathbf w^{(1)} = (6,3)$ since there are 6 inputs and 3 outputs, and for $\mathbf w^{(2)} = (4,1)$ again counting the $x^{(1)}_0$ node. Then we can write the following:
$$
x^{(1)}=w^{(1)T}x^{(0)}=(3,6)\times(6,1)=(3,1) \\
x^{(2)}=w^{(2)T}x^{(1)}=(1,4)\times(4,1)=(1,1)
$$
Note that we added the $x^{(1)}_0$ term. We can also represent this visually as follows
<img src="fwd.png" alt="Drawing" style="width: 550px;"/>
The number of operations are then defined by the dimension of the weight matrix where we have:
$$
\mathbf w^{(1)} = 6*3=18 \\
\mathbf w^{(2)} = 4*1 = 4
$$

Hence in total for forward propagation, i.e. for $w^{(l)}_{ij}x^{(l-1)}_i$ we have **22 operations**.  

For backward propagation we know that $\delta^{2}$ is of $(1,1)$ dimension since there is only one output node, and we already know from forward propagation that $x^{(1)}$ is $(4,1)$ dimension. Then for the final layer we have 
$$
\delta^{(2)}x^{(1)}=(1,1)\times(3,1)=(3,1)
$$
leading to **3 operations**.  
And lastly for the term $w^{(l)}_{ij}\delta^{(l)}_j$ we should have the same number of operations as in forward propagation as the weight matrix didn't change. Hence, when computing 
$$
w^{(2)}\delta^{(2)} = (4,1)\times(1,1)=(4,1)
$$
leading to **4 operations** and when computing
$$
w^{(1)}\delta^{(1)} = (6,3)\times(3,1)=(6,1)
$$
leading to **18 operations**

Hence, in total we have 22+3+4+18 = **47 operations**.  

Number of operations closest to the given potential answers is answer (d).

### Question 9
Let us call every node in a Neural Network a unit, whether that unit is an input variable or a neuron in one of the layers. Consider a Neural Network that has 10 input units (the constant $x^{(0)}_0$ is counted here as a unit), one output unit, and 36 hidden units (each $x^{(l)}_0$ is also counted as a unit). The hidden units can be arranged in any number of layers $l = 1, ... , L-1$, and each layer is fully connected to the layer above it.  

What is the minimum possible number of weights that such a network can have?

**Answer**: To obtain the minimum possible number of weights we need to *extend* the network as much as possible. Since each layer has the $x^{(l)}_0$ node, extending (making it deeper) the network we loose one connection at each layer $(l)$. Hence, the hidden layers should be arranged in such structure to have one unit node and one input/output node. Given 36 possible units that leads to 18 layers each with two units where one is $x^{(l)}_0$. So, given such combination we have the following:

$$
(10*1)+((36/2-1)*2)+(2*1) = 46
$$

where first multiplication term is for the input layer and first layer with one output node, the second term are 18 layers (but can't count the 1st since it does not accept inputs) minus 1 each with two weights and final term is the output node where we have two inputs and one output. Hence, the answer is (a).

### Question 10
What is the maximum possible number of weights that such a network can have?

Contrary to the previous question we should extend the network *vertically* rather than in depth (*horizontally*). Since we have one output node it does not benefit to have a complete oposite of the structure with minimum number of weights. We need a design to remedy this, but should not extend it more in depth as this will reduce the number of weights. So two layer combination should be optimal for maximization. Since there are 10 nodes in the input layer to obtain the maximal structure we should have more nodes in the 1st hidden layer than in the second one. Playing with some numbers comes to the following structure (if we try to increase layer 1 more it reduces the structure):
$$
10\times22\times14\times1
$$

which gives the following weight count:

$$
(10*(22-1))+(22*(14-1))+(14*1)=510
$$

The answer is (e).
